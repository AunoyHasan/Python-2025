### What is Matplotlib?
Matplotlib is a low level graph plotting library in python that serves as a visualization utility. 

```python
import matplotlib

print(matplotlib.__version__)
```

### Pyplot

```python
# Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

import matplotlib.pyplot as plt

import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([45, 25])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# Plotting without line

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
# Multiple Lines

xpoints = np.array([4, 6, 8, 10])
ypoints = np.array([10, 4, 3, 40])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# Default xpoints

ypoints = np.array([4, 8, 56, 78, 44])
plt.plot(ypoints)
plt.show()
```

### Markers
You can use the keyword argument marker to emphasize each point with a specified marker:

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
xpoints = np.array([4, 6, 8, 90, 77, 54, 89])
plt.plot(xpoints, marker = "*")
plt.show()
```

### Linestyle
You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = "o", linestyle = 'dotted')
plt.show()
```

```python

plt.plot(ypoints, marker = "o", linestyle = 'dashed')
plt.show()
```

```python

plt.plot(ypoints, marker = "o", linestyle = 'dotted', color ="r")
plt.show()
```

```python

plt.plot(ypoints, marker = "o", linestyle = 'dotted', color = "y", linewidth = '20.5')
plt.show()
```

### Matplotlib Labels and Title

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

### Matplotlib Adding Grid Lines

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

### Subplots

With the subplot() function you can draw multiple plots in one figure:

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)  # #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)  # #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x,y)

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

### Scatter Plot

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

### Bar Plot

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

### Histogram

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

```python
x = np.random.normal(170, 10, 250)

print(x)
```

```python
plt.hist(x)
plt.show() 
```

### Pie Charts

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```
